作业一：

安装并使用 requests、bs4 库，爬取猫眼电影的前 10 个电影名称、电影类型和上映时间，并以 UTF-8 字符集保存到 csv 格式的文件中。

```python
#构造头部，进行伪装
from fake_useragent import UserAgent
import requests
ua = UserAgent()
my_url = "https://maoyan.com/board" #请求的网址,猫眼电影排行榜，一页刚好10个
my_header = {"User-Agent":ua.random} #请求头
response = requests.get(url = my_url, headers = my_header) #请求网址
print(response.text) #响应体内容，其实也不太好用，多尝试几次才成功
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(response.text, 'lxml') 
data = soup.find_all('div', attrs={'class': 'movie-item-info'})
count = 1
data_csv = [['名次', '电影名称', '链接', '主演', '上映时间']]
for i in data:
    this_url = i.a['href']
    film_name = i.a['title']
    film_actor = i.find('p',class_ = 'star').text.strip().strip('主演：')
    film_releasetime = i.find('p',class_ = 'releasetime').text.strip('上映时间：')
    data_csv.append([count, film_name, 'https://maoyan.com'+this_url, film_actor, film_releasetime])
    count += 1
data_csv
```

```python
import unicodecsv as ucsv
with open('week-1-work-1.csv', 'wb') as f:
    w = ucsv.writer(f, encoding='utf-8-sig')
    w.writerows(data_csv)
import pandas as pd
data_show = pd.read_csv('week-1-work-1.csv')
data_show
```
